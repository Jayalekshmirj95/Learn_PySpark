# Introduction to Spark

* Spark was developed in Scala and Spark runs on JVM
* Because it runs on JVM, it lets you call into other Java based systems such as Hbase, HDFS and cassandra
* Spark runs locally or in a cluster (Hadoop YARN, Apache Mesos) or on-prem or on cloud (Amazon EC2, IBM Bluemix)
* Designed to work on distributed systems, but we can also use it on a single machine mode
* In order to access these modes, we use Spark Core API
* Spark Ecosystem consists of: 
    - Spark SQL and Dataframes - for batch processing
    - Spark Streaming - for real time operations
    - MLib - for machine learning operations
    - GraphX - to use Graph datastructure operations



# Spark Core API

Spark Core API
* contains the basic functionalities of Spark including task scheduling, memory management, fault-tolerance etc
    * Spark SQL and DataFrames
        * DF programming abstraction
        * complex analytics with SQL
    * Spark streaming
        * process real-time data
        * can handle both streaming and historical data
        * can use the same code for batch data and real-time data
    * MLib
        * high quality algos
    * GraphX
        * enables interaction with graph-structured data
        * introduces a new graph abstraction, the directed multigraph
        * use - visualizing social networks
        


Comparisons
* PySpark
    * Python wrapper around the spark core
* Pandas
    * great for tabular data
    * can handle millions of data
    * mature and feature rich
    * limited to a single machine
* Hadoop
    * distributed system
    * until a couple of years ago, Hadoop was the bigdata platform
    * Hadoop has a compute system called Map-Reduce and a file system called HDFS. They are closely integrated and difficult to run one without the other
    * Hadoop reads the data from and write to disk while Spark can do it in-memory. 
* Dask
    * Python library for parallel processing
    * Written in Python and only supports python
    * uses Pandas API and no SQL


# Spark Architecture

* Driver Program 
    * Sits on a node on the cluster
    * It maintains the details of the Spark Applications
    * It distributes the tasks and schedules work across different worker nodes
* Executors and Worker nodes
    * They carry out the tasks assigned by the driver
    * And reports back wiht a state of the process to the driver
* Cluster manager
    * Manages and provides the resource to the nodes
    * Types: Standalone, Kubernetes, Mesos, Hadoop YARN
    
Both driver and executors are processes which means they can exists in one machine if we run the application on Local(they just be threads) or different machines if we run the application on a cluster
